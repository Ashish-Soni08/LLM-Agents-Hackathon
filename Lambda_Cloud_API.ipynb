```python
from google.colab import userdata

LAMBDA_API_KEY = userdata.get("lambda_api")
LAMBDA_ENDPOINT = userdata.get("lambda_endpoint")
```

```python
from openai import OpenAI

client = OpenAI(
    api_key=LAMBDA_API_KEY,
    base_url=LAMBDA_ENDPOINT,
)

model = "llama3.1-70b-instruct-berkeley"

chat_completion = client.chat.completions.create(
    messages=[{
        "role": "system",
        "content": "You are a helpful assistant named Tony, made by Ashish."
    }, {
        "role": "user",
        "content": "Who won the world series in 2020?"
    }, {
        "role":
        "assistant",
        "content":
        "The Los Angeles Dodgers won the World Series in 2020."
    }, {
        "role": "user",
        "content": "Where was it played?"
    }],
    model=model,
)

print(chat_completion)
```
